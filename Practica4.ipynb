# Class

```python
class Salchipapa():
    # Class variable
    type_of_food = 'Junk Food'
    def __init__(self, name, price):
        self.name = name
        self.price = price
        self.ingredients = []
    def add_ingredients(self, ingredients):
        self.ingredients.append(ingredients)

ranchera_sencilla = Salchipapa('Ranchera Sencilla', 15000)
ranchera_especial = Salchipapa('Ranchera Especial', 20000)
print(ranchera_especial.type_of_food)
print(ranchera_sencilla.type_of_food)

ranchera_sencilla.add_ingredients(['Salchicha', 'Queso'])
print(ranchera_sencilla.ingredients)

print(ranchera_sencilla.price)
```

```python
from abc import ABC, abstractmethod

class AbstractEmployee(ABC):
    new_id = 1
    name = None
    company = 'Pepito'
    
    def __init__(self):
        self.id = AbstractEmployee.new_id
        AbstractEmployee.new_id +=1
        self._name = AbstractEmployee.name
        self._company = AbstractEmployee.company
        
    @abstractmethod
    def get_name(self):
        print('getter method called') 
        print(f'\nMy name is {self._name}')
        #return self._name
        
    def set_name(self, n):
        print("setter method called")
        self._name = n  
        
    def delete_name(self):
        print("deletter method called") 
        self._name = None
      
    def say_id(self):
        print(f'My ID is {self.id}')
    
    def get_id(self):
        return self.id
    
    def Dir(self):
        print(f'My company is {self._company}')
        
class Employee(AbstractEmployee):
    
    def __init__(self):
        super().__init__()
        
    def say_id(self):
        super().say_id()
        
    def get_name(self):
        super().get_name()
        
    def set_name(self, n):
        super().set_name(n)
        
    def delete_name(self):
        super().delete_name()

class User():
    
    def __init__(self, username, role):
        self.username = username
        self.role = role
        
    def say_user_info(self):
        print(f'User info: \nUsername:{self.username} Role:{self.role}')
        
class Admin(Employee, User):
    
    def __init__(self):
        super().__init__()
        User.__init__(self, self.id, 'Admin')
        
    def say_id(self):
        print('I am an Admin.')
        super().say_id()

class Manager(Admin):
    
    def __init__(self):
        super().__init__()
        User.__init__(self, self.id, 'Manager')
        
    def say_id(self):
        print('I am in charge.')
        super().say_id()

class Meeting():
    def __init__(self):
        self.attendees = []
        
    def __add__(self, employee):
        self.attendees.append(employee)
        print(f'Employee with ID {employee.get_id()} added succesfully')
        
    def __len__(self):
        return len(self.attendees)
        
e1 = Employee()
e2 = Employee()
e3 = Admin()
e4 = Manager()
e5 = User('5', 'User')

e1.say_id()
print('-'*30)
e2.say_id()
print('-'*30)
e3.say_id()
print('-'*30)
e4.say_id()
print('-'*30)

e3.say_user_info()
print('-'*30)
e4.say_user_info()
print('-'*30)
e5.say_user_info() 
print('-'*30)

meet = [Employee(), Admin(), Manager()]
for attend in meet:
    attend.say_id()
print('-'*30)

m1 = Meeting()
m1 + e1; m1 + e2; m1 + e3
print(len(m1))
print('-'*30)

e1.get_name()
print('-'*30)

e1.set_name('J')
print('-'*30)

e1.get_name()
print('-'*30)

e1.delete_name()
print('-'*30)

e1.get_name()
print('-'*30)

e1.Dir()
```

# Robot 

```python
class robot:
    type_of = 'Robot'
    
    def __init__(self, codigo, alias, creador, fecha):
        self._codigo = codigo
        self._alias = alias
        self._creador = creador
        self._fecha = fecha
    
    def get_code(self):
        print(f'\nCode: {self._codigo}')
        
    def get_alias(self):
        print(f'\nAlias: {self._alias}')
        
    def set_alias(self, n):
        print(f'El alias se ha cambiado a: {n}')
        self._alias = n
        
    def get_creador(self):
        print(f'\nCreador: {self._creador}')
        
    def get_fecha(self):
        print(f'\nFechar: {self._fecha}')
        
robot1 = robot('000', 'Pol', 'Yo', '03/01/2023')

robot1.get_code()
print('-'*30)

robot1.get_alias()
print('-'*30)

robot1.set_alias('J')
print('-'*30)

robot1.get_alias()
print('-'*30)

robot1.get_creador()
print('-'*30)

robot1.get_fecha()
print('-'*30)
```
